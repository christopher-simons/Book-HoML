```python
# Author: Chris Simons
# Date: 10 May 2020 

# Using the Normal Equation (page 108 on)

import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression


X = np.random.rand( 100, 1 )
y = 4 + 3 * X + np.random.rand( 100, 1 )

lin_reg = LinearRegression( )
lin_reg.fit( X, y )

print( "Linear Regression Intercept (bias) is :", lin_reg.intercept_ )
print( "Linear Regression Coefficient is :", lin_reg.coef_ )

y_prediction_lr = lin_reg.predict( X )

plt.plot( X, y_prediction_lr, color = 'green')
plt.scatter( X, y )
plt.show( )
```

```python
# using Stochastic Gradient Descent (page 117 on)
# can be quicker (but possibly more schochastic) than a Linear Regressor 

from sklearn.linear_model import SGDRegressor

# The following regressor runs for 50 epochs, starting with a learnign rate of 0.1 (eta0=0.1),
# and it does not use any regularisation ( penalty = None ).

sgd_reg = SGDRegressor( max_iter = 50, tol = 0.001, penalty = None, eta0 = 0.1 )

sgd_reg.fit( X, y.ravel( ) )

print( "SGD Regression Intercept (bias) is :", sgd_reg.intercept_ )
print( "SGD Regression Coefficient is :", sgd_reg.coef_ )

y_prediction_sgd = sgd_reg.predict( X )

plt.plot( X, y_prediction_sgd, color = 'red')
plt.scatter( X, y )
plt.show( )

```
